```python
n = 2
t = 1
m = 2
timetable = ['09:10', '09:09', '08:00']

```

```python
timetable = sorted(timetable)

#시간 + 분을 00:00으로부터 경과한 분 수로 변환 (as_min)
as_min = []
for time in timetable:
    hr = time[0:2]
    minute = time[3:5]
    as_min.append(60*int(hr)+int(minute))
    

# 시작 시간은 9:00 = 540분, 마지막 시간은 지금으로부터 t*(n-1) 분 이후 
now = 540
last = now + t*(n-1)
total_seat = n*m
before = 1440
single_seat = m

#마지막 버스 이전까지 자리 채우기
while total_seat > m:
    # 사람이 다 탔으면 break
    if len(as_min) == 0:
        break
    
    # 현재시간보다 먼저 온 사람이 있다면 사람을 태움
    # 남은 자리가 0개가 되면 현재시간을 다음 버스 시간으로 update
    elif as_min[0] <= now:
        as_min.remove(as_min[0])
        single_seat = single_seat - 1
        total_seat = total_seat -1
        if single_seat == 0:
            now = now + t
            single_seat = m
            n=n-1
    # 기다리는 사람이 없다면 자리가 남아 있어도 다음 버스 시간으로 update
    else:
        n = n-1
        now = now + t
        single_seat = m
        total_seat = n*m
        continue

#마지막 버스는 콘이 타야 하므로 같이 타는 사람들이 온 시간 정보를 함께 기록, 빈 자리가 1개일때까지 반복
last_group = []
for _ in range(m-1):
    if len(as_min) == 0:
        break
    before = as_min[0]
    # 이 경우 before = 콘이 안왔을 경우 마지막-1 번째로 버스에 탑승한 사람의 시간 (새치기 할 사람, case1) 
    if as_min[0] <= now:
        last_group.append(before)
        as_min.remove(before)
    # 이 경우 before = 막차보다 늦게 온 사람들 중 가장 빨리 온 사람 (case2)
    else:
        break
        
        


# 남은 대기 인원,같이타는 그룹, 마지막-1 번째로 버스 탑승한 사람의 탑승시간
print(as_min,last_group,before)

#  새치기 할 사람이 막차보다 늦으면 콘은 막차를 타고 감
if before>last:
    corn = last
    # 대신 1명이 정원이고 그게 첫차보다 빠르면 그 사람보다 1분 빨리오기
    if len(timetable)==1 and as_min[0]<=540:
        corn = as_min[0]-1


# 리스트에 남은 사람이 없으면 여유 자리가 있다는 뜻이므로 마지막으로 탄 사람 과 같은 시간 또는 막차를 타고 감
elif len(as_min) == 0:
    corn = max(last,before)


# asmin[0] = 콘이 안왔으면 마지막 자리에 앉았을 사람의 탑승시간
else:
    corn = as_min[0]
    # 
    if len(last_group) > 0:
        if corn >= max(last_group)-1: 
            corn = corn - 1

                
hr = corn//60
mn = corn%60
print(str(hr).zfill(2)+":"+str(mn).zfill(2))
```
