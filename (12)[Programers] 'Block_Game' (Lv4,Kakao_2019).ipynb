```python
# 1. 각 블럭 별 인덱스 추출 (dct)
# 2. 각 블럭 별 직사각형을 만들기 위한 빈칸의 인덱스 추출 (need_to_delete)
# 3. 빈칸의 위쪽의 모든 값이 0일경우 보드의 값을 0으로 변환
# 4. 지워지는 칸이 없을 때 까지 3을 반복하여 count
```

```python
def need_to_delete(block):
    row = [x[0] for x in block]
    col = [x[1] for x in block]
    
    # 조건 : 블럭들은 연속된 3개의 행 또는 열이 무조건 존재한다. 따라서 최소 + 최대 / 2 = 중간 블럭의 인덱스가 되며
    # 연속된 3개의 블럭이 행으로 있는지, 열로 있는지에 따라 구분
    if (max(row)+min(row))/2 in row:
        #연속된 3개의 블럭이 행으로 있는 경우
        #채워야 할 열은 전체 4개의 블럭을 합친 값에서 3*평균(무조건 3개가 있는 행의 값이 나온다)로 구할 수 있음 (c)
        
        c = sum(col) - 3*round(sum(col)/4)
        # 열에 해당하는 3가지 중 block에 포함된 값을 제거
        need = [[min(row),c],[min(row)+1,c],[min(row)+2,c]]
        for n in need:
            if n in block:
                need.remove(n)
                break
    else:
        # 반복
        r = sum(row) - 3*round(sum(row)/4) 
        need = [[r,min(col)],[r,min(col)+1],[r,min(col)+2]]
        for n in need:
            if n in block:
                need.remove(n)
                break
    return(need)

def solution(board):
    #1. 블럭 인덱스
    dct = {}
    for r, row in enumerate(board):
        for c, col in enumerate(row):
            if col != 0:
                try:
                    dct[col]
                except:
                    dct[col] = []
                dct[col].append([r,c])
    
    #2. 채워야 할 파트 인덱스
    need = {}
    for i in dct:
        need[i] = need_to_delete(dct[i])
    
    cnt = 0
    key = list(need.keys())
    while True:
        before = cnt
        for j in key:
            # 채워야 할 칸이 가로인가? 세로인가? 
            # need_to_delete 함수에서 더 효율적이게 코드를 짤 수 있을 것 같지만 나중에 고치겠습니다.
            if need[j][0][0] == need[j][1][0]:
                # 가로인 경우 2개의 열을 확인해야 한다.
                check1 = need[j][0][1]
                check2 = need[j][1][1]
                check_col1 = [x[check1] for x in board]
                check_col2 = [x[check2] for x in board]
                for k in range(0,need[j][1][0]+1):
                    if check_col1[k] != 0 or check_col2[k] != 0:
                        break
                else:
                    #for - else 문으로 for에서 break이 한번도 걸리지 않을 경우 보드에서 블럭을 없애준다 (0으로 바꿔준다.)
                    cnt = cnt + 1
                    delete = dct[j]
                    for d in delete:
                        board[d[0]][d[1]] = 0
                    key.remove(j)
            else:
                check = need[j][0][1]
                check_col = [x[check] for x in board]
                for k in range(0,need[j][1][0]+1):
                    if check_col[k] != 0:
                        break
                else:
                    cnt = cnt + 1
                    delete = dct[j]
                    for d in delete:
                        board[d[0]][d[1]] = 0
                        
                    key.remove(j)
        # 삭제된 블럭이 하나도 없으면 while문 종료
        if before == cnt:
            break

    answer = cnt
    return answer
```
